```python

import numpy as np 
import pandas as pd 
# import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'E:\cods\featuer scaling\wine-clustering.csv')

data = data[['Alcohol','Malic_Acid']]
data.head()
# print(type(data))
```

```python
import numpy as np

def feature_scaling(data):
    # Standardization
    mean = np.mean(data, axis=0)
    std = np.std(data, axis=0)
    standardized_data = (data - mean) / std
    
    # Min-Max Normalization
    min_val = np.min(data, axis=0)
    max_val = np.max(data, axis=0)
    normalized_data = (data - min_val) / (max_val - min_val)
    
    return np.round(standardized_data,3), np.round(normalized_data,3)

standardized_data, normalized_data = feature_scaling(data)
print('standardized_data:')
print(standardized_data)
print("Min-Max Normalization:")
print(normalized_data)
```

```python
print(type(normalized_data))
```

```python
datas = data.to_numpy()
```

```python
standardized = standardized_data.to_numpy()

normalized = normalized_data.to_numpy()
```

```python
fig, ax = plt.subplots(3, figsize=(6,14))
for a,d,l in zip(range(len(ax)),
               (datas, standardized, normalized),
               ('Primary data',
                'Standardized ',
                'Normalized')
                ):
    
    if a == 0:
        c = '#21918c'
    elif a == 1:
        c = '#fde725'
    else: 
        c = '#440154'
        
    ax[a].scatter(d[:, 0], d[:, 1], color = c,  s=50)
  
    ax[a].set_title(l)
    ax[a].set_xlabel('Alcohol')
    ax[a].set_ylabel('Malic Acid')
    ax[a].grid()
plt.tight_layout()
plt.show()
```
